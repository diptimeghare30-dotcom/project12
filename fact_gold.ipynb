```python
%sql
USE CATALOG dm_pro2;

CREATE OR REPLACE TABLE gold.fact_daily_sales AS
SELECT
    o.OrderDate,
    SUM(o.TotalAmount) AS DailyRevenue,
    SUM(o.Quantity) AS UnitsSold,
    COUNT(DISTINCT o.OrderID) AS OrdersCount
FROM silver.orders o
GROUP BY o.OrderDate;


CREATE OR REPLACE TABLE gold.fact_product_performance AS
SELECT
    p.ProductID,
    p.ProductName,
    p.Category,
    p.Supplier,
    SUM(o.Quantity) AS TotalUnitsSold,
    SUM(o.TotalAmount) AS TotalRevenue,
    SUM(r.RefundAmount) AS TotalRefunds,
    (SUM(r.RefundAmount) / SUM(o.TotalAmount)) AS RefundRate
FROM silver.products p
LEFT JOIN silver.orders o
  ON p.ProductID = o.ProductID
LEFT JOIN silver.returns r
  ON o.OrderID = r.OrderID
GROUP BY p.ProductID, p.ProductName, p.Category, p.Supplier;

```
