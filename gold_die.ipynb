```python
from pyspark.sql.functions import *
from pyspark.sql.types import *


catalog_name = 'dm_pro2'


# Top categories by revenue potential
df_silver = spark.table(f'{catalog_name}.silver.products')
df_category_revenue = df_silver.groupBy("Category").sum("Price")
#df_category_revenue.show()
df_silver.write.format("delta") \
    .mode("overwrite") \
    .option("mergeSchema", "true") \
    .saveAsTable(f"{catalog_name}.gold.category_revenue")


# Top loyalty customers
df_silver_customer = spark.table(f'{catalog_name}.silver.customers')
df_top_customers = df_silver_customer.filter(col("LoyaltyPoints") >= 250)
#df_top_customers.show()
df_silver.write.format("delta") \
    .mode("overwrite") \
    .option("mergeSchema", "true") \
    .saveAsTable(f"{catalog_name}.gold.top_customers")
```
